```python
#Ans 4)
#In Python, a generator is a special type of function which does not return a single value, isteeat , it return an iterator object with a sequence of values. It allows you to generate a series of values on the fly, rather than generating them all at once and storing them in memory.
#The yield keyword is used to return a value form the generator function and pause the execution of the function until the next value is requested.

def test1(n):
    i = 1
    while i <=n:
        yield i
        i +=1
```

```python
a = test1(8)
```

```python
for i in a:
    print (i)
```
